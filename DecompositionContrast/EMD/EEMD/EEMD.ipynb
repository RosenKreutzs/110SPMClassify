# EEMD的实现和效果展示
- 添加白噪音的目的，是为了同化真实噪音，以便在集合平均时同时削弱真实噪音和白噪音的影响
- 我的个人理解：[EEMD](./EEMD.md)；

## step1:设置EEMD的函数

```python
import numpy as np  
from PyEMD import EMD
"""
eemd函数：(用于对信号进行EEMD分解)
- num_imf ：假设我们想要IMF的个数  
- num_ensemble ：集合中信号的数量  
- noise_std：添加到数据中的白噪声的标准差 
"""
def eemd(data, num_imf, num_ensemble, noise_std):  
    imfs_ensemble = []  
    for _ in range(num_ensemble):  
        noise = np.random.normal(0, noise_std, size=data.shape)  
        emd = EMD()  
        IMFs=emd.emd(data + noise)  
        residual = data - np.sum(IMFs, axis=0)
        imfs_ensemble.append(IMFs)  
  
    # 对所有IMF进行平均  
    avg_imfs = [np.mean([IMFs[i] for IMFs in imfs_ensemble], axis=0) for i in range(num_imf)]  
    return avg_imfs
"""
reconstruct_signal函数：(EEMD分解得到的IMFs还原为信号）
"""
def reconstruct_signal(imfs, residue=None):  
    # 如果没有提供残余项，则只使用IMFs进行重构  
    if residue is None:  
        reconstructed_signal = np.sum(imfs, axis=0)  
    else:  
        reconstructed_signal = np.sum(imfs, axis=0) + residue  
    return reconstructed_signal 
```

## step2:对原始信号进行EEMD操作，获得IMF组；

```python
import numpy as np  
from PyEMD import EMD

  # 参数  
num_imf = 7  # 假设我们想要5个IMF  
num_ensemble = 100  # 集合中信号的数量  
noise_std = 0.2  # 添加到数据中的白噪声的标准差
signal=X_train_first_12[0]

IMFs = eemd(signal, num_imf, num_ensemble, noise_std)   
residual = signal - np.sum(IMFs, axis=0)
reconstructed_signal = reconstruct_signal(IMFs,residual)
```

## step3:显示IMF组和原始信号和重构信号；
- 每个IMF都代表了原始信号的某种固有振动模式信息；
- IMF包含的信息可通过[筛选](../../ModelsContrast/110spmemdlstmvibration.ipynb)后，以达到去噪的效果；之后再重构；

```python
plt.plot(signal, 'r')
plt.title("原始信号")
plt.show()

plt.figure(figsize=(15,10))
for num in range(num_imf):
    plt.subplot(num_imf+1, 1, num+2)
    plt.plot(IMFs[num])
    plt.title("IMF "+str(num+1))
plt.show()

plt.plot(residual, 'r')
plt.title("残差信号")
plt.show()

plt.plot(reconstructed_signal, 'r')
plt.title("重构信号")
plt.show()
```
