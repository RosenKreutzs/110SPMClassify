## step1:设置ICEEMDAN的函数

```python
from PyEMD import EMD

def ICEEMDAN(signal, num_imfs, num_ensembles=100, noise_strength=0.2, max_iter=100):
    """
    通过ICEEMDAN对信号进行分解为多个IMF
    
    参数:
    signal (array-like): 输入的一维信号
    num_imfs (int): 需要提取的IMF数量
    num_ensembles (int): 加入噪声的集合数目
    noise_strength (float): 噪声强度
    max_iter (int): 最大迭代次数
    
    返回:
    imfs (array-like): 分解得到的IMFs
    """
    N = len(signal)
    imfs = np.zeros((num_imfs, N))
    
    # 初始化EMD
    emd = EMD()

    # 计算第一阶IMF
    noise = noise_strength * np.random.randn(num_ensembles, N)
    signal_plus_noise = np.array([signal + noise[i] for i in range(num_ensembles)])
    imf1_ensemble = np.array([emd.emd(signal_plus_noise[i], max_imf=1) for i in range(num_ensembles)])
    imfs[0] = np.mean(imf1_ensemble[:, 0, :], axis=0)
    
    # 计算后续的IMF
    for i in range(1, num_imfs):
        residue = signal - np.sum(imfs[:i], axis=0)
        imf_ensemble = []
        
        for j in range(num_ensembles):
            noise = noise_strength * np.random.randn(N)
            imf_ensemble.append(emd.emd(residue + noise, max_imf=1)[0])
        
        imfs[i] = np.mean(imf_ensemble, axis=0)
    
    return imfs
```

## step2:对原始信号进行ICEEMDAN操作，获得IMF组；

```python
import numpy as np
from PyEMD import CEEMDAN
# 假设你已经有一个信号
signal = X_train_first_12[0]  # 将长度为2048的信号替换为你的信号


# 使用ICEEMDAN对信号进行分解
IMFs = ICEEMDAN(signal, num_imfs=6)


# 计算残差
residual = signal - np.sum(IMFs, axis=0)

# 信号重构
reconstructed_signal = np.sum(IMFs, axis=0)

print("IMFs shape:", IMFs.shape)
print("Residual shape:", residual.shape)
print("Reconstructed signal shape:", reconstructed_signal.shape)
```
