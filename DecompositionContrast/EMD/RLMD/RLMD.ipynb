## step1:设置RLMD的函数

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

def compute_local_mean_and_envelope(x):
    """
    计算局部均值和包络线（上包络和下包络）。
    """
    peaks_max, _ = signal.find_peaks(x)
    peaks_min, _ = signal.find_peaks(-x)
    
    # 线性插值计算上包络和下包络
    upper_envelope = np.interp(np.arange(len(x)), peaks_max, x[peaks_max])
    lower_envelope = np.interp(np.arange(len(x)), peaks_min, x[peaks_min])
    
    local_mean = (upper_envelope + lower_envelope) / 2
    return local_mean, upper_envelope, lower_envelope

def RLMD(x, max_imfs=10, max_iters=100, stopping_threshold=0.001):
    """
    RLMD 分解
    """
    imfs = []
    residual = x.copy()
    
    for _ in range(max_imfs):
        h = residual.copy()
        
        for _ in range(max_iters):
            # 计算局部均值和包络线
            local_mean, upper_env, lower_env = compute_local_mean_and_envelope(h)
            h1 = h - local_mean
            
            # 检查收敛条件
            if np.linalg.norm(h - h1) < stopping_threshold:
                break
            h = h1
        
        # 提取一个IMF
        imfs.append(h)
        residual = residual - h
        
        # 停止条件：当残差为单调或近似零时，结束
        if np.all(np.diff(residual) > 0) or np.all(np.diff(residual) < 0) or np.linalg.norm(residual) < stopping_threshold:
            break
    
    return imfs, residual

def reconstruct_signal(imfs, residual):
    """
    使用IMFs和残差重构信号
    """
    return np.sum(imfs, axis=0) + residual
```

## step2:对原始信号进行RLMD操作，获得IMF组；

```python
import numpy as np
from PyEMD import CEEMDAN
import pandas as pd
# 假设你已经有一个信号
x = X_train_first_12[0]  # 将长度为2048的信号替换为你的信号


# 使用RLMD对信号进行分解
IMFs,residual =RLMD(x)
IMFs=pd.Series(IMFs)
residual=pd.Series(residual)

# # 计算残差
# residual = signal - np.sum(IMFs, axis=0)

# 信号重构
reconstructed_signal = np.sum(IMFs, axis=0)+ residual

print("IMFs shape:", IMFs.shape)
print("Residual shape:", residual.shape)
print("Reconstructed signal shape:", reconstructed_signal.shape)
```
